```python
# list of apis from dataset
l_apis = set(df_api.columns)
# l_apis
```

## count the num of apks invoking funcitons for each plugin

```python
d_plugins_cnt = {k:[0] *len(d_plugins_all) for k, _ in d_plugins_1.items()}
d_plugins_cnt
```

```python
# count the num of apks invoking funcitons for each plugin
import functools
for i, d_plugins in enumerate(d_plugins_all):
    for plugin, apis in d_plugins.items():
    #     api_cnt = 0 
        df_temp = deepcopy(df_api)
        l_apis_in = []
        for api in apis:
            if api in l_apis:
                l_apis_in.append(api)
    #             api_cnt += (df_api[api] == 1).sum()
        if l_apis_in:
            print(f'{plugin}: {apis}')
            print(f'feature matched in extracted apis: {l_apis_in}')
            conditions = [df_temp[api]==0 for api in l_apis_in]
            df_filter = df_temp[functools.reduce(np.logical_and, conditions)]
            api_cnt = df_temp.shape[0] - df_filter.shape[0]
            d_plugins_cnt[plugin][i] += api_cnt
            print(f"Found {api_cnt}/{df_temp.shape[0]} samples use the provided methods for the plugin {plugin}")
            print("\n")
    print(d_plugins_cnt)
        
```
