# Different Feature Set

## prune permission fetures

```python
df_permissions_standard = pd.read_csv("../db/permission_levels_updated.csv", skipinitialspace=True)
df_permissions_standard
```

```python
# a list of columns names for plugins with usage, declaration, permison respectively
l_plugins = list(df_pro.columns[:47])
l_plugins
```

```python
l_permission_standard = list(Counter(df_permissions_standard["permission"]).keys())
print(len(l_permission_standard))
l_permission_standard
```

```python
# update df with plugins and standard permission only
l_init = ["apk_name"] + l_plugins
l_cols_df_pro = list(df_pro.columns)
df_update = df[l_init]
l_not_detected_permission = []
for col in l_permission_standard:
    if col in l_cols_df_pro:
        df_update[col] = df_pro[col]
    else:
        l_not_detected_permission.append(col)
        df_update[col] = 0
print(f"{len(l_not_detected_permission)}/{len(l_permission_standard)} permission not detected in the dataset")
print(l_not_detected_permission)
```

```python
df_update
```

## Add numders of permissions detected for each level in each APK

```python
ls_permissions_standard = Counter(df_permissions_standard["level"])
ls_permissions_standard
```

```python
# creata dict level to store total number of level for each apk
rows, cols = df_update.shape
d_level = {key: [0] * rows for key in ls_permissions_standard.keys()}
print(len(list(d_level.values())[0]))
d_level
```

```python
# a dict of permission level
d_permission_level = pd.Series(df_permissions_standard.level.values,index=df_permissions_standard.permission).to_dict()
d_permission_level
```

```python
for index, row in df_update.iterrows():
#     print(index)
    for p, l in d_permission_level.items():
        if row[p] == 1:
            d_level[l][index] += 1
d_level
```

```python
df_level = pd.DataFrame(d_level)
df_level
```

```python
df_update_level = pd.concat([df_update, df_level.reindex(df_update.index)], axis=1)
df_update_level
```

```python
df_update_level["class"] = df_pro["class"]
df_update_level
```

## output csv

```python
# ROC Curve for the classification models

from sklearn.metrics import roc_auc_score, roc_curve, auc
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from numpy import interp

# Setting the parameters for the ROC Curve
plt.rcParams['figure.figsize'] = [10,8]
plt.style.use("bmh")
color = cm.rainbow(np.linspace(0, 1, len(model_ls)))
plt.title("ROC Curve", fontsize = 15)
plt.xlabel("Specificity (False Positive Rate)", fontsize = 15)
plt.ylabel("Sensitivity (True Positive Rate)", fontsize = 15)
labels = list()


# Plot ROC Curve
for c, (name, y_pred) in enumerate(y_pred_ls.items()):
    
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color[c])
    labels.append('{} (AUC={:.3f})'.format(name, roc_auc))
  
plt.gca().legend(labels, loc='lower right', frameon=True)    
plt.plot([0,1],[0,1], linestyle='--', color='black')
plt.show()
```
